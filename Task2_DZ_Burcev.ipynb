```python
#Приводим колонки датафрейма к одному типу данных

print('Previous Datatypes\n', df.dtypes, sep='')
for column in df.columns:
    if df[column].dtype != 'float64':
        try:
            df[column] = df[column].astype(float)
        except ValueError:
            print(f"Cannot convert values in column '{column}' to float")

print('New Datatypes\n', df.dtypes, sep='')
```

```python
#Проверяем индекс, чтобы он соответствовал порядку строк

df = df.sort_index(ascending=False)
```

```python
#Проверяем наличие пропусков

missing_values = df.isnull().sum()
print(missing_values)
```

```python
#для удобства перемименновываем столбцы
import pandas as pd

# Создаем словарь с новыми названиями столбцов
new_column_names = {
    'rpdRegionCd': 'region',
    'objAddr': 'address',
    'objReady100PercDt': 'date',
    'objSquareLiving': 'price',
    'objElemLivingCnt': 'object',
    'objFloorMin': 'FloorMin',
    'objFloorMax': 'FloorMax',
    'objPriceAVG': 'PriceAVG'
    
}

# Используем метод rename() для переименования столбцов
df = df.rename(columns=new_column_names)
```

```python
#проверяем изминения по столбцам
df.head()
```

```python
df.shape
```

```python
#Средние цены за квадратный метр по регионам

#группируем по региону и сортируем по возрастанию по средней стоимости квартиры 

mean_price=df.groupby('region',as_index=False)\
             .aggregate({'price':'mean'})\
             .sort_values('price')\
             .rename(columns={'price':'mean_price'})

#результат самые дорогие квартиры в Москве и СПБ  
mean_price=mean_price.reset_index()
del mean_price['index']
mean_price
```

```python
#Рассчитывем объемы вводимого жилья по регионам и по годам
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Группируем данные по регионам и годам и суммируем объемы жилья
df_grouped = df.groupby(['region', 'date']).sum()['object']

# Преобразуем результат в новый датафрейм
df_volume = df_grouped.reset_index()

# Строим график
plt.figure(figsize=(30, 25))
for region in df_volume['region'].unique():
    region_data = df_volume[df_volume['region'] == region]
    plt.plot(region_data['date'], region_data['object'], label=region)
plt.xlabel('Дата')
plt.ylabel('Кол-во обьектов')
plt.title('Объемы вводимого жилья по регионам и по годам')
plt.legend()
plt.show()

#Самый большой обьем вводмиого жилья в регинах приходится на 2023-2024г
```

```python
# Сравниваем этажность домов, количество квартир (обьектов) и цену за квадратный метр в разных регионах.

comparison = df.groupby('region').agg({
    'FloorMax': 'mean',
    'FloorMin': 'mean',
    'object': 'sum',
    'PriceAVG': 'sum',
    'price': 'mean'
})

# Вывод результатов
print(comparison)
```

```python
#Сравниваем этажность домов по регионам
import matplotlib.pyplot as plt

# Создаем графика
plt.figure(figsize=(10, 6))
plt.plot(comparison.index, comparison['FloorMax'], label='Средняя макс. этажность')
plt.plot(comparison.index, comparison['FloorMin'], label='Средняя мин. этажность')
plt.xlabel('Регион')
plt.ylabel('Этажность')
plt.title('Сравнение средней макс. и мин. этажности по регионам')
plt.legend()
plt.xticks(rotation=45)

# Отображение графика
plt.show()

#Ср.максимальная этажность приходится на центральные регионы страны и Чеченский регион
#Ср.минимальная этажность приходится также на центральные регионы страны 
```

```python
import matplotlib.pyplot as plt

# Создание графика
plt.figure(figsize=(10, 8))
plt.plot(comparison.index, comparison['object'], label='Обьем строищехся квартир в регионах')
plt.xlabel('Регион')
plt.ylabel('Количество')
plt.title('Обьем строищехся квартир в регионах')
plt.legend()
plt.xticks(rotation=40)

# Отображение графика
plt.show()

#Самый большой обьем вводмиого жилья в регинах приходится Москву и СПБ
```

```python
import matplotlib.pyplot as plt

# Создание графика
plt.figure(figsize=(10, 8))
plt.plot(comparison.index, comparison['PriceAVG'], label='Средняя стоимость квартиры в регионах')
plt.xlabel('Регион')
plt.ylabel('Цена')
plt.title('Средняя стоимость квартиры в регионах')
plt.legend()
plt.xticks(rotation=45)

# Отображение графика
plt.show()

#Самый большая средняя стоимость жилья в регинах приходится Москву и СПБ
```

```python
#Связь между средней макс. этажностью и ценой за квадратный метр по регионам
plt.figure(figsize=(10, 6))
plt.scatter(comparison['FloorMax'], comparison['price'])
plt.xlabel('Средняя макс. этажность')
plt.ylabel('Цена за квадратный метр')
plt.title('Связь между средней макс. этажностью и ценой за квадратный метр по регионам')
plt.show()

#На графики в явном виде отражается история чем выше этаж тем дороже кв.м
```

```python
#Распределение количества объектов по регионам
plt.figure(figsize=(10, 6))
plt.hist(comparison['object'], bins=10)
plt.xlabel('Количество объектов')
plt.ylabel('Частота')
plt.title('Распределение количества объектов по регионам')

# Отображение графиков
plt.show()
```

```python
# boxplot для разброса 'object', 'PriceAVG' и 'price'
plt.boxplot([comparison['object'], comparison['PriceAVG'], comparison['price']])
plt.xticks([1, 2, 3], ['object', 'PriceAVG', 'price'])
plt.xlabel('Показатель')
plt.ylabel('Значение')
plt.title('Разброс показателей')
plt.show()
```

```python
#График рассеяния для связи между минимальным и максимальным кол-вом этажей
plt.scatter(comparison['FloorMax'], comparison['FloorMin'])
plt.xlabel('FloorMax')
plt.ylabel('FloorMin')
plt.title('Связь между FloorMax и FloorMin')
plt.show()

```

```python
#Гистограмма для распределения максимального кол-во этажей по регионам
plt.hist(comparison['FloorMax'], bins=10)
plt.xlabel('FloorMax')
plt.ylabel('Частота')
plt.title('Распределение FloorMax')
plt.show()

#На графике видно что в среднем самые популярные этажи 10-11
```
